```python
import pandas as pd


# reading the database
data = pd.read_csv("/content/Heart.csv")

# printing the top 10 rows
display(data.head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Scatter plot
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Age', y='Chol', hue='AHD', data=data)
plt.title('Scatter Plot of Age vs Cholesterol with Heart Disease Status')
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(6, 4))
sns.countplot(x='ChestPain', hue='AHD', data=data)
plt.title('Bar Plot of Chest Pain Types with Heart Disease Status')
plt.show()
```

```python

# Box plot
plt.figure(figsize=(6, 4))
sns.boxplot(x='AHD', y='MaxHR', data=data)
plt.title('Box Plot of Max Heart Rate with Heart Disease Status')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(6, 4 ))
sns.histplot(data['RestBP'])
plt.title('Histogram of Resting Blood Pressure')
plt.xlabel('Resting Blood Pressure')
plt.ylabel('Frequency')
plt.show()
```
